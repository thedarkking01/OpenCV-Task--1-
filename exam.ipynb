```python
import cv2
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('sample.png')
```

```python
plt.imshow(image)
plt.title('Sample CV Image')
plt.show()
```

```python
# 3. Find the height and width of the image
height, width, channels = image.shape
print(f"Height: {height}, Width: {width}, Channels: {channels}")
```

```python
#2.Convert Colored Image to Grayscale
grayscale_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.imshow(grayscale_image, cmap='gray')
plt.title("Grayscale Image")
plt.show()
```

```python
# 3. Generate Negative Image of Input
negative_image = 255 - grayscale_image
plt.imshow(negative_image)
plt.title("Negative Image")
plt.show()
```

```python
# flip vertically
mirror_x = cv2.flip(image, 0)
plt.imshow(mirror_x)
plt.title('Flip Vertical')
plt.show()
```

```python
# flip horizontally
mirror_y = cv2.flip(image, 1)
plt.imshow(mirror_y)
plt.title('Flip Horizontal')
plt.show()
```

```python
# 3. Draw a Rectangle
top_left = (100, 150)     # Top-left corner of the rectangle
bottom_right = (400, 300) # Bottom-right corner of the rectangle
color = (0, 0, 255)       # Red color in BGR
thickness = 2             # Line thickness
cv2.rectangle(image, top_left, bottom_right, color, thickness)
```
